# Read in datasets

```python
param_dfs = []

param_names = [
    "num_particles",
    "inertia",
    "cognition",
    "social"
]

int_columns = [
    "particles",
    "function",
    "epochs"
]

for p in param_names:
    param_dfs.append(pd.read_csv(f"pso-{p}.csv", encoding="utf-8").dropna())
    
    param_dfs[-1].name = p

    for ic in int_columns:
        param_dfs[-1][ic] = param_dfs[-1][ic].astype("int32")

print(param_names[0])
display(param_dfs[0])
```
